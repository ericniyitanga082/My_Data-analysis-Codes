```python
# Load the Child and Mother dataset
df_child = pd.read_spss('2_CFSVA18_DB_ChildAndMother_Full_Annex_201904_NISR.sav')

# Quick check
print("Shape:", df_child.shape)       # rows x columns
print("Columns:", df_child.columns.tolist())
df_child.head()                        # see first 5 rows
```

```python
# Load the Household dataset
df_house = pd.read_spss('1_CFSVA18_DB_HouseholdQues_Full_Annex_201904_NISR.sav')

# Quick check
print("Shape:", df_house.shape)
print("Columns:", df_house.columns.tolist())
df_house.head()
```

```python
# Run this to see your most important columns
key_cols = [
    # Target variable (what you want to predict)
    'HAZ', 'WAZ', 'WHZ',          # nutrition scores

    # Child factors
    'UrbanRural',                  # urban or rural
    'S0_C_Prov',                   # province
    'WI_cat',                      # wealth category

    # Will come from household after merge
    'VHHSize',                     # household size
    'FinalWeight',                 # survey weight
]

# Check which ones exist
found = [c for c in key_cols if c in df_child.columns or c in df_house.columns]
print("Key columns found:", found)

# Preview them
df_child[['UrbanRural','S0_C_Prov','WI_cat']].value_counts().head(10)
```

```python
# Step 1: Find ALL nutrition/anthropometry columns
nutri_cols = [c for c in df_child.columns
              if any(x in c.lower() for x in
              ['haz','waz','whz','muac','stunt','wast','underw','nutri','zscor','anthro'])]

print(f"Found {len(nutri_cols)} nutrition columns:\n")
for col in nutri_cols:
    non_null = df_child[col].notna().sum()
    print(f"  {col:30s} | non-null: {non_null} | sample: {df_child[col].dropna().unique()[:4]}")
```

```python
# ============================================
# STEP 4: Find more child-level feature columns
# ============================================
# Search for age, sex, feeding, health columns
extra_cols = [c for c in df_child.columns
              if any(x in c.lower() for x in
              ['age','sex','gender','feed','breast','diet','water',
               'toilet','health','vacc','diarr','fever'])]

print(f"Found {len(extra_cols)} potential feature columns:")
for col in extra_cols:
    print(f"  {col:35s} | nulls: {df_child[col].isna().sum()}")
```

```python
# Print ALL columns with their index number
print("ALL COLUMNS IN df_child:\n")
for i, col in enumerate(df_child.columns):
    print(f"  {i:3d}. {col}")
```

```python
# ============================================
# CHECK ALL OUR CHOSEN FEATURES
# ============================================

features_to_check = [
    # Location
    'S0_C_Prov', 'UrbanRural', 'S0_E_Livezone',

    # Poverty
    'WI_cat', 'Ubudehe', 'FS_final',

    # Household
    'VHHSize',

    # Child
    'S13_02', 'Age_cat_child',

    # Mother
    'S13_01_4',

    # Feeding (6-59 months)
    'MAD_FBF4_CDDS', 'MAD_FBF4_CMMFS', 'MAD_FBF4_CMAD',

    # Diet
    'Tot_MDDW', 'Meat_MDDW', 'Eggs_MDDW', 'GreenVeg_MDDW',

    # Health
    'S13_06', 'S13_07'
]

print("=" * 60)
for col in features_to_check:
    print(f"\n📌 {col}")
    print(f"   Type    : {df_child[col].dtype}")
    print(f"   Nulls   : {df_child[col].isna().sum()}")
    print(f"   Values  : {df_child[col].dropna().unique()[:6]}")
print("=" * 60)
```

```python
# Check S13_02 surrounding columns for sex
sex_cols = df_child.iloc[:, 155:170]  # columns around S13_02
for col in sex_cols.columns:
    print(f"{col}: {df_child[col].dropna().unique()[:5]}")
```

```python
# Check anthropometry section for sex column
print("Checking columns 98-125:\n")
for col in df_child.columns[98:125]:
    unique_vals = df_child[col].dropna().unique()[:5]
    nulls = df_child[col].isna().sum()
    print(f"{col:20s} | nulls: {nulls:4d} | values: {unique_vals}")
```

```python
# Check Section 14 columns (22-97)
print("Checking S14 columns:\n")
for col in df_child.columns[22:55]:
    unique_vals = df_child[col].dropna().unique()[:5]
    nulls = df_child[col].isna().sum()
    print(f"{col:20s} | nulls: {nulls:4d} | values: {unique_vals}")
```

```python
# Check the remaining S14 columns we are unsure about
cols_to_check = ['S14_05', 'S14_06', 'S14_07',
                 'S14_09', 'S14_10', 'S14_12',
                 'S14_13', 'S14_14']

print("Checking uncertain S14 columns:\n")
for col in cols_to_check:
    nulls = df_child[col].isna().sum()
    vals = df_child[col].dropna().unique()[:5]
    count = df_child[col].value_counts().head(3).to_dict()
    print(f"📌 {col}")
    print(f"   Nulls : {nulls}")
    print(f"   Values: {vals}")
    print(f"   Counts: {count}")
    print()
```

```python
# ============================================
# CREATE OUR WORKING DATASET
# ============================================

selected_columns = [
    # Target
    'Global_Stunting',

    # Location
    'S0_C_Prov', 'UrbanRural', 'S0_E_Livezone',

    # Economic Status
    'WI_cat', 'Ubudehe', 'FS_final',

    # Child
    'S14_01_5', 'ENA_DOBM', 'S14_07', 'S14_09',
    'S14_10', 'S14_13', 'S14_14',

    # Mother
    'S13_01_3', 'S13_01_5', 'S13_03', 'S13_05',

    # Diet & Feeding
    'S14_06', 'Tot_MDDW', 'Meat_MDDW',
    'Eggs_MDDW', 'GreenVeg_MDDW'
]

# Create working dataset
df = df_child[selected_columns].copy()

print("✅ Working dataset created!")
print(f"\nShape: {df.shape}")
print(f"Rows    : {df.shape[0]} children")
print(f"Columns : {df.shape[1]} variables")
```

```python
# ============================================
# RENAME COLUMNS TO MEANINGFUL NAMES
# ============================================

df = df.rename(columns={
    # Target
    'Global_Stunting'  : 'stunting',

    # Location
    'S0_C_Prov'        : 'province',
    'UrbanRural'       : 'urban_rural',
    'S0_E_Livezone'    : 'livelihood_zone',

    # Economic
    'WI_cat'           : 'wealth_index',
    'Ubudehe'          : 'ubudehe_category',
    'FS_final'         : 'food_security',

    # Child
    'S14_01_5'         : 'child_sex',
    'ENA_DOBM'         : 'child_age_months',
    'S14_07'           : 'vaccinated',
    'S14_09'           : 'had_diarrhea',
    'S14_10'           : 'vitamin_a',
    'S14_13'           : 'mosquito_net',
    'S14_14'           : 'deworming',

    # Mother
    'S13_01_3'         : 'mother_age',
    'S13_01_5'         : 'mother_education',
    'S13_03'           : 'breastfeeding',
    'S13_05'           : 'mother_preg_lact',

    # Diet
    'S14_06'           : 'meals_per_day',
    'Tot_MDDW'         : 'diet_diversity_score',
    'Meat_MDDW'        : 'ate_meat',
    'Eggs_MDDW'        : 'ate_eggs',
    'GreenVeg_MDDW'    : 'ate_vegetables'
})

print("✅ Columns renamed successfully!")
print("\nNew column names:")
for col in df.columns:
    print(f"   {col}")
```
